```python
'''
Normal random disadv
------------------------------------------
Random Generators (Why Use Them)
Legacy np.random.* uses a global state, causing issues in multithreading or parallel code where calls interfere. np.random.Generator (via default_rng()) fixes this:

Thread-safe: Each generator has independent state, avoiding race conditions.
Reproducible: Seed once per generator: rng = np.random.default_rng(42). Same seed yields identical sequences across runs.
Faster PCG64: Default algorithm outperforms older Mersenne Twister for vectorized operations.
Controlled streams: Create multiple generators for separate tasks without global state pollution.
'''
```

```python
import numpy as np
rng = np.random.default_rng(42)  # Seeded generator
print(rng.random(3))  # [0.773956, 0.643865, 0.177394]
print(rng.integers(0, 10, 3))   # Controlled integers
```
