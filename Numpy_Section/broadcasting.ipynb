```python
import numpy as np
```

```python
arr1=np.array([1,2,3])
arr2=np.array([4,5,6])
print(arr1.shape,arr2.shape)
arr1+arr2
```

```python
arr1=np.array([1,2,3]) # becomes
# [[1,2,3]
#  [1,2,3]
#  [1,2,3]]
arr2=np.array([[4],
               [5],
               [6]])
# [[4,4,4],
#  [5,5,5],
#  [6,6,6]]
arr1+arr2
```

```python
# failing condition
arr1=np.array([1,2,3])
print(arr1.shape)
arr2=np.array([4,5,6,7])
print(arr2.shape)

arr1+arr2
# rules are from right to left the axis should be either equal or 1, here this condition failing 3!=4 and there is no 1
```
